```python
import warnings
warnings.filterwarnings(action='ignore')

#20년 달력 프레임 만들기
#12,1,2 :겨울 / 3,4,5 : 봄 / 6,7,8 : 여름 / 9,10,11 : 가을

import datetime

calendar = pd.DataFrame()

date = pd.date_range('2020-01-01', '2020-12-31')
year = date.year
month = date.month
day = date.day
weekday = date.day_name()
weekend = np.ones(len(date)) # 초기값 선언 (int)

calendar['date'] = date
calendar['year'] = year
calendar['month'] = month
calendar['day'] = day
calendar['weekday'] = weekday
calendar['weekend'] = weekend
calendar['season'] = "" 

for i in range(len(date)):
    if weekday[i] == 'Saturday' or weekday[i] == 'Sunday':
        weekend[i] = 1
    else:
        weekend[i] = 0

for i in range(len(date)):
    if month[i] == 12 or month[i] == 1 or month[i] == 2 :
        calendar['season'][i] = "winter"
    elif month[i] == 3 or month[i] == 4 or month[i] == 5 :
        calendar['season'][i] = "spring"
    elif month[i] == 6 or month[i] == 7 or month[i] == 8 :
        calendar['season'][i] = "summer"
    else : 
        calendar['season'][i] = "fall"
```
